# A Use of Dictionaries

```
import numpy as np

HALF=.5
MOTOR_SPEEDS = {
    "q": (HALF, 1), "w": (1, 1), "e": (1, HALF),
    "a": (-1, 1), "d": (0, 0), "d": (1, -1),
    "z": (-HALF, -1), "x": (-1, -1), "c": (1, HALF),
}

print(10 * np.array(MOTOR_SPEEDS['q']))
```
